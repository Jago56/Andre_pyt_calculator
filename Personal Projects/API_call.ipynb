Code fetches public GitHub user data, converts it to a Python dictionary, and prints it **— (foundation of all API work in Python.)**

import requests  
Loads the tool that lets Python talk to APIs.

url = "https://api.github.com/users/octocat"  
This is the API endpoint (like a phone number).

response = requests.get(url)  
Python sends a GET request to GitHub.

data = response.json()  
Converts the JSON response into a Python dictionary.

print(data)  
Shows the data returned by the API.

Error handling included

```python
import requests

def call_api(url):
    try:
    # Try to send the request
        response = requests.get(url, timeout=10)

        # Raise an error if the status code is 4xx or 5xx
        response.raise_for_status()

        # Convert JSON to Python dictionary
        return response.json()
###Error: The request timed out.###
    except requests.exceptions.Timeout:
        print("Error: The request timed out.")

###HTTP Error: 404 Client Error: Not Found###
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP Error: {http_err}")

###Error: Could not connect to the server.
    except requests.exceptions.ConnectionError:
        print("Error: Could not connect to the server.")

    except requests.exceptions.RequestException as err:
        print(f"Unexpected Error: {err}")

# Example usage
url = "https://api.github.com/users/octocat"
data = call_api(url)

print(data)
```
