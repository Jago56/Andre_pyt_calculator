```python
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment, PatternFill, Border, Side

import os

# Create a new Excel workbook
workbook = Workbook()

# Select the active sheet
sheet = workbook.active

# Rename the sheet (optional)
sheet.title = "DataSheet"

sheet["A1"] = "Name"
sheet["B1"] = "Age"

header_font = Font(bold=True)
center_align = Alignment(horizontal="center")

sheet["A1"].font = header_font
sheet["B1"].font = header_font
sheet["B1"].alignment = center_align

# -----------------------------
# Write data rows (corrected)
# -----------------------------
data = [
    ("Andre", 35),
    ("John", 29),
    ("Vino", 46),
    ("Chadd", 42),
    ("Jago", 2)
]

row_start = 2
for name, age in data:
    sheet[f"A{row_start}"] = name
    sheet[f"B{row_start}"] = age
    sheet[f"B{row_start}"].alignment = center_align
    row_start += 1

# -----------------------------
# Add borders to all cells
# -----------------------------
thin_border = Border(
    left=Side(style="thin"),
    right=Side(style="thin"),
    top=Side(style="thin"),
    bottom=Side(style="thin")
)

for row in sheet.iter_rows(min_row=1, max_row=1+len(data), min_col=1, max_col=2):
    for cell in row:
        cell.border = thin_border

# -----------------------------
# Adjust column widths
# -----------------------------
sheet.column_dimensions["A"].width = 18
sheet.column_dimensions["B"].width = 10

# -----------------------------
# Save the Excel file
# -----------------------------
filename = "Andre_practice.xlsx"
workbook.save(filename)

# Get the full absolute path
full_path = os.path.abspath(filename)

print("Excel file created successfully!")
print("Full file path:", full_path)

# Download in Google Colab
from google.colab import files
files.download(filename)


# Write data into cells
sheet["A1"] = "Name"
sheet["B1"] = "    Age"

sheet["A2"] = "Andre"
sheet["B2"] = 35

sheet["A3"] = "John"
sheet["B3"] = 29

sheet["A4"] = "Vino"
sheet["B4"] = 46


sheet["A5"] = "Chadd"
sheet['B5'] = 42


sheet['A6'] = "Jago"
sheet["B6"] = 2

# File name
filename = "Andre_practice.xlsx"

# Save the Excel file
workbook.save(filename)

# Get the full absolute path
full_path = os.path.abspath(filename)

print("Excel file created successfully!")
print("Full file path:", full_path)
from google.colab import files
files.download("Andre_practice.xlsx")
```
